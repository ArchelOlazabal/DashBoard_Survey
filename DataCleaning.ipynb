### Removing Empty Columns

```python
rows = original_data.isna().sum()
rows
```

```python
columns_to_remove = ['Email', 'Browser', 'OS', 'City', 'Country', 'Referrer'] # The "Email" column was removed due to all the values said "anonymous" therefore has no significant value.
df1 = original_data.drop(columns_to_remove, axis=1)
rows = df1.isna().sum()
rows
```

### There are columns that have "Other (Please Specify):" which is not ideal when creating the dashboard. The code below cleans all these columns that have similar issues/wording.

```python
def remove_characters(cell):
    if isinstance(cell, str):  #checks if cell is string
        return cell.replace('Other (Please Specify):', '')
    return cell 

df1 = df1.applymap(remove_characters)
df1
```

### Putting all characters in lowercase form for consititancy

```python
def lower_characters(cell):
    if isinstance(cell, str):  #checks if cell is string
        return cell.lower()
    return cell 

df1 = df1.applymap(lower_characters)
df1
```

### Cleaning and normalizing the Q5 column whihc describes their coding language that they like.

```python
df1['Q5 - Favorite Programming Language'].unique()
```

Removing the "other:" string from the data

```python
df1['Q5 - Favorite Programming Language'] = df1['Q5 - Favorite Programming Language'].str.replace('other:', '')


df1['Q5 - Favorite Programming Language'].unique()
```

If the answer given contains the word sql then the answer will be cateogrized as "sql"

```python
df1['Q5 - Favorite Programming Language'] = df1['Q5 - Favorite Programming Language'].apply(lambda x : 'sql' if 'sql' in x else x)
df1['Q5 - Favorite Programming Language'] = df1['Q5 - Favorite Programming Language'].str.strip()
df1['Q5 - Favorite Programming Language'].unique()
```

```python
# Cleaning the answer to the character change to unique value to NA
answer_change = {
    "dont require" : "na",
    "i don't know any": "na",
    "unknown":"na",
    "i don't use programming in my role":"na",
    "i currently do not work with programming languages yet":"na",
    "none":"na",
    "just started learning":"na",
    "mainly use excel": "excel",
    "none at the moment":"na",
    "i do analysis and create presentations based on datasets provided by others":"na",
    "none at the moment":"na",
    "na at the moment":"na"
}

def cleaning_answer(text):
    for misspelled, correct in answer_change.items():
        text = text.replace(misspelled, correct)
    return text

df1['Q5 - Favorite Programming Language'] = df1['Q5 - Favorite Programming Language'].apply(cleaning_answer)
df1['Q5 - Favorite Programming Language'].unique()
```

### Correcting and standardizing Q11 column which describes where the surveyer lives.

```python
# removing any empty spaces around the word
df1['Q11 - Which Country do you live in?'] = df1['Q11 - Which Country do you live in?'].str.strip()
df1['Q11 - Which Country do you live in?'].unique()
```

```python
# Cleaning the answer by correcting the misspelling
misspelled_to_correct = {
    'perú': 'peru',
    'uae': 'united arab emirates',
    'portugsl': 'portugal',
    'irel': 'ireland',
    'ire': 'ireland',
    'brazik': 'brazil',
    'africa (nigeria)': 'nigeria',
    'fin': 'finland',
    'austr': 'australia',
    'leba': 'lebanon',
    'hong kong': 'china',
    'kenua': 'kenya',
    'sg': 'singapore',
    'aisa': 'asia',
    'argentine': 'argentina',
    'uzb': 'uzbekistan',
    'other (please specify)': ''
}

def correct_misspellings(text):
    for misspelled, correct in misspelled_to_correct.items():
        text = text.replace(misspelled, correct)
    return text

df1['Q11 - Which Country do you live in?'] = df1['Q11 - Which Country do you live in?'].apply(correct_misspellings)
df1['Q11 - Which Country do you live in?'] = df1['Q11 - Which Country do you live in?'].str.capitalize()
df1['Q11 - Which Country do you live in?'].unique()
```

### Cleaning Q1 column that describe the persons job title.

```python
df1['Q1 - Which Title Best Fits your Current Role?'].unique()
```

```python
text_correction = {
    'other (please specify)': 'other',
    'does a social media analyst count?': 'social media analyst',
    'manager of a team of data analysts': 'data analyst manager',
    'reporting adm': 'reporting administrator',
    'i work with data tools and can create simple dashboards but i am not a data scientist':'data analyst',
    'pmo': 'pmo (project management office) manager',
    'student working as a data analyst intern ': 'data analyst intern',
    'dba': 'database administrator',
    'rf engineer': 'radio frequency engineer'
}

def correction(text):
    for misspelled, correct in text_correction.items():
        text = text.replace(misspelled, correct)
    return text

df1['Q1 - Which Title Best Fits your Current Role?'] = df1['Q1 - Which Title Best Fits your Current Role?'].apply(correction)
df1['Q1 - Which Title Best Fits your Current Role?'] = df1['Q1 - Which Title Best Fits your Current Role?'].str.strip()
df1['Q1 - Which Title Best Fits your Current Role?'].unique()
```

### Since Q3 asks about the surveyer salary most gave ranges so the average yearly salary was taken.

```python
def average_income(text):
    try:
        lower, upper = map(lambda x: int(x.strip('k')), text.split('-'))
        return (upper + lower) / 2
    except ValueError:
        result = text.strip('k+')
        return int(result)
    
df1['Q3 - Current Yearly Salary (in USD)'] = df1['Q3 - Current Yearly Salary (in USD)'].apply(average_income)
df1['Q3 - Current Yearly Salary (in USD)'].unique()
```
